```python
n=15
a=[]
for i in range(1,n+1):
  if (i%3==0)and(i%5==0):
    a.append("FizzBuzz")
  elif(i%5==0):
    a.append("Buzz")
  elif(i%3==0):
    a.append("Fizz")
  else:
    a.append(str(i))
print(a)
```

```python
nums = [2,7,11,15]
target = 9
a=[]
for i in range(len(nums)-1):
  for j in range(i+1,len(nums)):
    if nums[i]+nums[j]==target:
      a.append(i)
      a.append(j)
      break
```
